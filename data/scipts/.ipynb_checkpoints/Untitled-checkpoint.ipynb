```python
import os
import unidecode
import json
import copy
```

```python
data_path = r'/Users/anujraaj/mythic/first-proto/data'
links_path = r'/Users/anujraaj/mythic/first-proto/data/links'
```

```python
links = []
with open(links_path) as f:
    links = f.readlines()
for (idx, link) in enumerate(links):
    links[idx] = link.rstrip("\n")
len(links)
```

```python
with open(os.path.join(data_path, "data.json"), "r") as read_file:
    data = json.load(read_file)
# for item in data:
#         item["character2"] = item["file_name"].split(".")[0].rstrip().split("_")[-1]
len(data)
```

```python
def get_url(s):
        opts = []
        for link in links:
                s2 = unidecode.unidecode(s.replace("_", ""))
                link2 = copy.deepcopy(link)
                link2 = link2.replace("-", "")
#                 print(s2, link)
                if link2.find(s2)!= -1 : 
                        opts.append(link)
        return opts
```

```python
for item in data:
    temp2 = copy.deepcopy(item["file_name"])
    temp = temp2.split(".")[0].rstrip().replace("_", "")
#     print(temp)
    url = get_url(temp)
    if temp == "54va":
        url = url[:1]
    if(len(url)!=1):
        print("BIG LUND")
#     url = get_url(item["file_name"].split(".")[0].rstrip())
#     url = get_url(item["file_name"])
#     print (item["character2"], url, len(url))
#     print(url, temp, item["character2"], item["file_name"])
    item["url"] = url
```
